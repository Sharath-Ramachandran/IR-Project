```python
import csv
import pandas as pd
import numpy as np
import re
X=[]
Y=[]
X1=[]
X2=[]
Y1=[]
regex = re.compile(r'[\t]')
with open('outfiles/merged_input_file_TrainB.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        row[0]=row[0].replace('"',"")
        row[0]=regex.sub(" ",row[0])
        li = list(row[0].split(" "))
        li = [i for i in li if len(i) >= 1]
        li = list(map(float, li))     
        X.append(li)
y=label
print(X[:2])
#X1=sum(X, [])
print(len(y))
print(len(X))
#print(X)

for i in range(3834):
    Y.append((list(y.iloc[i])))
    
Y2=[]
for i in range(3834):
    y2=[]
    for x in Y[i]:
        y2.append(float(x))
        Y2.append(y2)
print(Y2[:5])
#print(X2[:5])
#print(Y[:5])
y1=np.array(Y2)
x1=np.array(X)
```

```python
predict_TestB=[]
with open('outfiles/merged_input_file_TestB.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        row[0]=regex.sub(" ",row[0])
        li = list(row[0].split(" "))
        li = [i for i in li if len(i) >= 1]
        li = list(map(float, li))
        predict_TestB.append(li)
print(len(predict_TestB))
```
