```python
# import csv
import pandas as pd
df=pd.DataFrame(columns=["Tweet_index","Label","Tweet_Text"]);
txt_file=r"SemEval2018-T3-train-taskB_emoji_ironyHashtags.txt"
#csv_file=r"data11.csv"
i=0
with open(txt_file) as f:
    for line in f:
        #print(line)
        if i==0:
            i=i+1
            continue
        #print(line)
        df.loc[i]=[text for text in line.split("\t")]
        i=i+1

#df=pd.read_csv('input_data_TrainA.csv',sep='\t')
#print(df)
tweets=df["Tweet_Text"]
label=df["Label"]
print(len(tweets))
```

```python
import csv
import pandas as pd
import numpy as np
import re
X=[]
Y=[]
X1=[]
X2=[]
Y1=[]
regex = re.compile(r'[\t]')
with open('outfiles/merged_input_file_TrainB.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        row[0]=row[0].replace('"',"")
        row[0]=regex.sub(" ",row[0])
        li = list(row[0].split(" "))
        li = [i for i in li if len(i) >= 1]
        li = list(map(float, li))     
        X.append(li)
y=label
print(X[:2])
#X1=sum(X, [])
print(len(y))
print(len(X))
#print(X)

for i in range(3834):
    Y.append((list(y.iloc[i])))
    
Y2=[]
for i in range(3834):
    y2=[]
    for x in Y[i]:
        y2.append(float(x))
        Y2.append(y2)
print(Y2[:5])
#print(X2[:5])
#print(Y[:5])
y1=np.array(Y2)
x1=np.array(X)
```

```python
predict_TestB=[]
with open('outfiles/merged_input_file_TestB.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        row[0]=regex.sub(" ",row[0])
        li = list(row[0].split(" "))
        li = [i for i in li if len(i) >= 1]
        li = list(map(float, li))
        predict_TestB.append(li)
```

```python
import keras
from keras.optimizers import SGD , Adam
from keras.models import Sequential
from keras.layers import Dense
def buildmodel():
    print("Now we build the model")
    model = Sequential()
    model.add(Dense(60, activation='relu', input_dim=90))
    model.add(Dense(1, activation='softmax'))
    sgd = SGD(lr=0.0001)# decay=1e-6, momentum=0.9, nesterov=True)
    model.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
    print("We finish building the model")
    return model

model = buildmodel()
model.fit(x1,y1, epochs=41, batch_size=64)
```
