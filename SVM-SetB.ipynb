```python
# import csv
import pandas as pd
df=pd.DataFrame(columns=["Tweet_index","Label","Tweet_Text"]);
txt_file=r"SemEval2018-T3-train-taskB_emoji_ironyHashtags.txt"
#csv_file=r"data11.csv"
i=0
with open(txt_file) as f:
    for line in f:
        #print(line)
        if i==0:
            i=i+1
            continue
        #print(line)
        df.loc[i]=[text for text in line.split("\t")]
        i=i+1

#df=pd.read_csv('input_data_TrainA.csv',sep='\t')
#print(df)
tweets=df["Tweet_Text"]
label=df["Label"]
print(len(tweets))
```

```python
import emoji
import re
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk import ne_chunk
input_txt=""
for line in tweets:
    #print(line)
    de_emojize=emoji.demojize(line)
    commonurl_str = re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', 'http://someurl', de_emojize)
    replace_username=re.sub('(?<=^|(?<=[^a-zA-Z0-9-_\.]))@([A-Za-z]+[A-Za-z0-9-_]+)','@someuser',commonurl_str)
  #  puncutation_flooding=re.sub(r'['+string.punctuation+']+', ' ',replace_username)
    input_txt+=replace_username
print(input_txt)
input_tokenized=nltk.word_tokenize(input_txt)
#print(input_tokenized)
input_posTag=nltk.pos_tag(input_tokenized)
#print(input_posTag)
```

```python
import csv
import pandas as pd
import numpy as np
X=[]
regex = re.compile(r'[\t]')
with open('outfiles/merged_input_file_TrainB.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        #print(row[0])
        #print(type(row[0]))
        row[0]=row[0].replace('"',"")
        row[0]=regex.sub(" ",row[0])
        #print(row[0]
        li = list(row[0].split(" "))
        li = [i for i in li if len(i) >= 1]
        #print(li)
        #print(type(li))
        li = list(map(float, li))
        #print(li)
        X.append(li)
y=label
print(len(y))
print(len(X))
print(len(X[0]))
```

```python
a = np.asarray(X)
print(1/(len(a)*a.std()))
```

```python
import numpy as np
from sklearn.svm import SVC

clf = SVC(gamma='auto')
#for training data
clf.fit(X, y) 

SVC(C=1.0, cache_size=200, class_weight=None, coef0=0.0,
    decision_function_shape='ovr', gamma=0.01, kernel='rbf',
    max_iter=-1, probability=False, random_state=None, shrinking=True,
    tol=0.001, verbose=False)
```

```python
predict_TestA=[]
with open('outfiles/merged_input_file_TestB.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        row[0]=regex.sub(" ",row[0])
        li = list(row[0].split(" "))
        li = [i for i in li if len(i) >= 1]
        li = list(map(float, li))
        predict_TestA.append(li)
#print(predict_TestA)
```

```python
for i in range(len(predict_TestA)):
    clf.predict([predict_TestA[i]])
```

```python
output=[]
for i in range(len(predict_TestA)):
    output.append(clf.predict([predict_TestA[i]]).tolist())
    #print(type(value.tolist()))
print(output)
```
